```python
import os
from ultralytics import YOLO
import cv2

# üîπ Direct video path (you can paste your own)
video_path = "/Users/anantasati/Desktop/Movie on 10-04-25 at 5.45‚ÄØAM.mov"

# Load YOLO model
model_path = '/Users/anantasati/Desktop/runs/detect/train5/weights/best.pt'
model = YOLO(model_path)

# Output path: same directory as the YOLO run folder
output_dir = os.path.dirname(os.path.dirname(model_path))  # 'runs/detect/train'
video_name = os.path.splitext(os.path.basename(video_path))[0]
video_path_out = os.path.join(output_dir, f'{video_name}_out.mp4')

# Open input video
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape

# Prepare video writer
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'mp4v'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

threshold = 0.2

while ret:
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result
        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"\n‚úÖ Output video saved at:\n{video_path_out}")
```
